```python
import pandas as pd
df = pd.read_csv('output_lattes\curriculos_processados.csv')
df.head()
```

```python
df.drop(columns=['cidade_nascimento',
'codigo_instituicao',
'codigo_curso',
'codigo_area_curso',
'curso_grande_area',
'curso_area',
'grande_area_formacao',
'area_formacao'], inplace=True)
```

```python
df.info()
```

```python
df = df.rename(columns={
    'grande_area_definitivo': 'grande_area',
    'area_definitivo': 'area',
    'instituicao_uf': 'uf_instituicao',
    'instituicao_regiao': 'regiao_instituicao',
    'instituicao_pais': 'pais_instituicao'
})
```

```python
colunas_ordenadas = [
    'numero_identificador', 'genero', 'data_atualizacao',
    'uf_nascimento', 'regiao_nascimento', 'pais_nascimento',
    'tipo_formacao', 'status_curso', 'ano_inicio', 'ano_conclusao',
    'grande_area', 'area', 'flag_bolsa',
    'uf_instituicao', 'regiao_instituicao', 'pais_instituicao'
]
```

```python
df = df[colunas_ordenadas]
df.info()
```

```python
df['flag_bolsa'] = df['flag_bolsa'].apply(lambda x: True if str(x).lower() == 'sim' else False)
df['data_atualizacao'] = pd.to_datetime(df['data_atualizacao'], format='%d%m%Y', errors='coerce')
df['ano_inicio'] = df['ano_inicio'].astype('Int64')
df['ano_conclusao'] = df['ano_conclusao'].astype('Int64')
df.info()
```
